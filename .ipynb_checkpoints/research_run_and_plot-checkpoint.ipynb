### Calculate metrics for comparison

```python
time_res = "two_month"
num_clu = 10

turb_info = pd.read_csv('data/wind_data/DK/turb_info_test.csv')

unc = pd.read_csv('data/results/raw/'+str(year_test)+'_unc_cf.csv', parse_dates=['time'])
cor = pd.read_csv('data/results/raw/'+str(year_test)+'_'+time_res+'_'+str(num_clu)+'_cor_cf.csv', parse_dates=['time'])
obs = pd.read_csv('data/wind_data/DK/obs_cf_test.csv', parse_dates=['time'])

unc_year = unc.groupby(pd.Grouper(key='time',freq='Y')).mean().transpose().reset_index()
cor_year = cor.groupby(pd.Grouper(key='time',freq='Y')).mean().transpose().reset_index()
obs_year = obs.groupby(pd.Grouper(key='time',freq='Y')).mean().transpose().reset_index()

unc_year.columns = ["ID", "unc"]
cor_year.columns = ["ID", "cor"]
obs_year.columns = ["ID", "obs"]

all_cf = obs_year.merge(unc_year,on='ID').merge(cor_year,on='ID')
all_cf['ID'] = all_cf['ID'].astype('int')
all_cf = all_cf.merge(turb_info[['ID','lat','lon']],on='ID')
```

```python
unc_month = unc.groupby(pd.Grouper(key='time',freq='M')).mean().transpose().mean().values
cor_month = cor.groupby(pd.Grouper(key='time',freq='M')).mean().transpose().mean().values
obs_month = obs.groupby(pd.Grouper(key='time',freq='M')).mean().transpose().mean().values

plt.plot(range(1,13), unc_month*100, label='Uncorrected')
plt.plot(range(1,13), cor_month*100, label='Corrected')
plt.plot(range(1,13), obs_month*100, label='Observation')

plt.xlabel('Month')
plt.ylabel('Capacity Factor %')
plt.legend();
```

```python
param = 'obs' # CAN ALSO VIEW: 'uncorr', corrected'

fig = px.scatter_geo(all_cf, lat='lat',lon='lon', 
                     color = param,
                     hover_data=[param],facet_col_wrap=3,
                    color_continuous_scale='hot_r', range_color=(0,1)) #color='cluster',
              # other colors: sunsetdark, hot_r,plotly3_r
fig.update_geos(
    resolution=50, 
#     showland=True, landcolor="lavender",
    center=dict(lon=9.6, lat=56),
    lataxis_range=[54.8, 58.5], lonaxis_range=[7,15],
    projection_type='mercator')
fig.update_traces(marker=dict(size=3.5))
fig.update_layout(title = 'Denmark Wind Farms', title_x=0.5, width = 900, height = 900)
fig.show()
```

### Plotting for the paper

```python


sns.set_theme()
sns.set_style("white")
```

```python
g = sns.catplot(
    x='num_clu',
    y='MAE',
    hue = 'time_res', 
    data = metrics, 
    kind = 'point',
    errorbar = None,
    legend_out = False
)

g.set_axis_labels("No. of Clusters", "Mean Absolute Error (MAE)")
g.set_titles("{col_name}")
g._legend.set_title('Time Resolution')
```

```python
g = sns.catplot(
    x='num_clu',
    y='RMSE',
    hue = 'time_res', 
    data = metrics, 
    kind = 'point',
    errorbar = None,
    legend_out = False
)

g.set_axis_labels("No. of Clusters", "Root Mean Square Error (RMSE)")
g.set_titles("{col_name}")
g._legend.set_title('Time Resolution')
```
