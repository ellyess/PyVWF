```python
import vwf.VWF_Model as model
import time
import xarray as xr
import numpy as np
import pandas as pd

from vwf.extras import calculate_metrics_all

mode = 'merra2' # 'era5' or 'merra2'
method_list = ['method_1', 'method_2']
test_year = 2020
time_res_list = ['year', 'season', 'two_month', 'month'] 
# cluster_list = [1, 3, 10, 15, 50]
cluster_list = [1,2,3,5,10,25,50,100,200]
```

```python
if mode == 'merra2':
    year_star = 2017
    year_end = 2019
    method = 'method_1'
    test_year = 2020
    time_res = 'year'
    
    vwf_model = model.VWF(mode, method)
    vwf_model.load_files(year_star, year_end, test_year)
    vwf_model.test(time_res)
    
else:
    for method in method_list:

        if method == 'method_1':
            year_star = 2017
            year_end = 2019

        else:
            year_star = 2016
            year_end = 2019   

        vwf_model = model.VWF(mode, method)
        vwf_model.load_files(year_star, year_end, test_year)

        for num_clu in cluster_list:

            vwf_model.train(num_clu)

            for time_res in time_res_list:

                vwf_model.test(time_res)

    calculate_metrics_all(mode, method_list, test_year, cluster_list, time_res_list)
```
