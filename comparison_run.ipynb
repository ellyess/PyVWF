# VWF Model: Running multiple scenarios

This notebook exists for research purposes to easily run and compare all the various variations of this model:
- How the reanalysis data is processed into wind speed via Method 1 or Method 2.
- How spatial (no. of clusters) and temporal (from yearly to monthly) 

This code can take a long time to run as training needs to occur dependent on number of clusters.

```python
import vwf.VWF_Model as model
import time
import xarray as xr
import numpy as np
import pandas as pd

mode = 'era5'
method_list = ['method_1', 'method_2']
test_year = 2020
time_res_list = ['year', 'season', 'two_month', 'month'] 
# cluster_list = [1, 3, 10, 15, 50]
cluster_list = [1,2,3,5,10,25,50,100,200]
```

```python
for method in method_list:

    if method == 'method_1':
        year_star = 2017
        year_end = 2019

    else:
        year_star = 2016
        year_end = 2019   

    vwf_model = model.VWF(mode, method)
    vwf_model.load_files(year_star, year_end, test_year)

    for num_clu in cluster_list:

        vwf_model.train(num_clu)

        for time_res in time_res_list:

            vwf_model.test(time_res)
```
